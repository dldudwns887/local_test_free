```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd

# 붓꽃 데이터셋 불러오기
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
iris_df.head()
```

```python
# 히스토그램 시각화: 꽃잎 길이 (Petal length)
plt.figure(figsize=(8, 6))
plt.hist(iris_df["petal length (cm)"], bins=15, color="skyblue", edgecolor="black", alpha=0.7)
plt.title("Histogram of Petal Length", fontsize=16)
plt.xlabel("Petal Length (cm)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

```python
# Seaborn을 사용한 히스토그램: Petal Length
plt.figure(figsize=(8, 6))
sns.histplot(iris_df["petal length (cm)"], bins=15, kde=True, color="orange")
plt.title("Histogram of Petal Length with KDE", fontsize=16)
plt.xlabel("Petal Length (cm)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

```python
data1 = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
data2 = [2, 3, 3, 4, 4, 5, 5, 5, 5, 5]

plt.hist([data1, data2], bins=5, color=["blue", "orange"], label=["Data 1", "Data 2"], stacked=True)
plt.legend()
plt.show()
```

```python
import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=20, cumulative=True, color="purple", edgecolor="black")
plt.title("Cumulative Histogram")
plt.xlabel("Values")
plt.ylabel("Cumulative Frequency")
plt.show()
```

```python
plt.hist(data, color = 'green', alpha = 0.5, bins = 100, range = [-6, 6], edgecolor = 'black')
plt.show()
```
