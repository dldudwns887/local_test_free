```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

# 5차원 데이터 생성 (예제 데이터)
np.random.seed(42)
data = np.random.rand(100, 5)  # 100개의 샘플, 5개의 차원
df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5'])

# PCA로 차원 축소 (5D -> 3D)
pca = PCA(n_components=3)
data_3d = pca.fit_transform(df)

# 3D 시각화
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# 데이터 포인트 플롯
ax.scatter(data_3d[:, 0], data_3d[:, 1], data_3d[:, 2], c=data[:, 0], cmap='viridis', s=50)

# 축 레이블 설정
ax.set_title("3D Visualization of 5D Data (PCA)", fontsize=14)
ax.set_xlabel("Principal Component 1", fontsize=12)
ax.set_ylabel("Principal Component 2", fontsize=12)
ax.set_zlabel("Principal Component 3", fontsize=12)

plt.show()
```

```python
import plotly.express as px

# 3D 시각화 (대화형)
df_3d = pd.DataFrame(data_3d, columns=['PC1', 'PC2', 'PC3'])
df_3d['Color'] = data[:, 0]  # 첫 번째 특징을 색상에 매핑

fig = px.scatter_3d(df_3d, x='PC1', y='PC2', z='PC3', color='Color', title="3D Visualization with Plotly")
fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 생성
x = np.linspace(-5, 5, 100)  # X축 범위
y = np.linspace(-5, 5, 100)  # Y축 범위
X, Y = np.meshgrid(x, y)  # 메쉬 그리드 생성
Z = np.sin(np.sqrt(X**2 + Y**2))  # Z값: 높이 (예: 함수 값)

# 3D 등고선도
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# 3D 등고선도 그리기
contour = ax.contour3D(X, Y, Z, levels=15, cmap="viridis")
ax.set_title("3D 등고선도 예제")
ax.set_xlabel("X축")
ax.set_ylabel("Y축")
ax.set_zlabel("Z축 (높이 값)")

plt.show()
```
