```python
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 불러오기
iris = sns.load_dataset("iris")  # 붓꽃 데이터
titanic = sns.load_dataset("titanic")  # 타이타닉 데이터

# 시각화
plt.rc('font', family='Malgun Gothic')  # 한글 폰트 설정 (Windows 환경)
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

# 막대그래프 그리기
iris.sepal_length[:20].plot(kind='bar', rot=0)
plt.title("Iris 데이터: Sepal Length Top 20", fontsize=14)
plt.xlabel("Index", fontsize=12)
plt.ylabel("Sepal Length", fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# 꺾은선 그래프
plt.figure(figsize=(10, 6))

# 첫 번째 그래프
plt.plot(x, y1, color='blue', linestyle='-', linewidth=2, label='sin(x)', marker='o', markersize=5)

# 두 번째 그래프
plt.plot(x, y2, color='red', linestyle='--', linewidth=2, label='cos(x)', marker='s', markersize=5)

# 그래프 설정
plt.title("Sine and Cosine Functions", fontsize=16)
plt.xlabel("X Values", fontsize=12)
plt.ylabel("Y Values", fontsize=12)
plt.legend(fontsize=10)
plt.grid(True)
plt.show()
```
