```python
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd


# 엑셀 파일 경로
file_path = "C:\\Users\\dldud\\Downloads\\[60MinTide]202501_DT_0002.xls"

# 엑셀 파일을 읽어서 데이터프레임 생성
df = pd.read_excel(
    io=file_path,
    sheet_name="202501",  # 시트 이름
    header=3  # 첫 번째 행을 컬럼 이름으로 사용
)

print(df.head())
```

```python
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

# '시간'을 datetime 형식으로 변환
df["시간"] = pd.to_datetime(df["시간"])

# 그래프 그리기
plt.figure(figsize=(10, 5))
plt.plot(df["시간"], df["예측 조위"], marker=".", linestyle="-", label="예측 조위")
plt.title("시간별 예측 조위 변화")
plt.xlabel("시간")
plt.ylabel("예측 조위")

plt.legend()
plt.grid()
plt.tight_layout()  # 레이아웃 조정
plt.show()
```
